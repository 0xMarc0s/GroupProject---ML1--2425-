```python
def bin_age(age):
    if age < 21:
        return '<=21'
    if age < 26:
        return '(21, 26]'
    if age < 31:
        return '(26, 31]'
    if age < 36:
        return '(31, 36]'
    if age < 46:
        return '(36, 46]'
    if age < 51:
        return '(46, 51]'
    return '>51'
```

```python
order = [
    '<=21',
    '(21, 26]',
    '(26, 31]',
    '(31, 36]',
    '(36, 46]',
    '(46, 51]',
    '>51',
]
```

```python
hash_obesity_inverted = {
 1: 'Normal_Weight',
 2: 'Overweight_Level_I',
 3:'Overweight_Level_II',
 4:'Obesity_Type_I',
 5:'Insufficient_Weight',
 6:'Obesity_Type_II',
 7:'Obesity_Type_III'
 }

```

```python
df['obese_level'] = df['obese_level'].replace(hash_obesity_inverted)
```

```python
df['age_class'] = df['age'].apply(bin_age)
```

```python
fig, ax = plt.subplots()
sns.countplot(data=df, x='age_class', hue='obese_level', ax=ax, palette='Blues', order=order)
fig.suptitle('Obesity Level by Age Class')
ax.set_xlabel('Age Class by Interval')
```
